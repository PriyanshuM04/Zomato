# Zomato Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Zomato data .csv")
```

```python
df
```

#### Converting the data type of rate to float and remove the denominator

```python
def handle_rate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)

df['rate'] = df['rate'].apply(handle_rate)
```

```python
df.head()
```

```python
df.info()
```

<b>Conclusion</b>: There are NO NULL values

#### Type of Resturant

```python
sns.countplot(x = df['listed_in(type)'])
plt.xlabel("Type of Resturant")
```

<b>Conclusion</b> : Dining Resturants are preferred by a larger number of customers

```python
grouped_data = df.groupby("listed_in(type)")["votes"].sum()
result = pd.DataFrame({"votes" : grouped_data})
plt.plot(result, c = "green", marker = "o")
plt.xlabel("Type of Resturant", c = "red", size = 15)
plt.ylabel("Votes", c = "red", size = 15)
```

```python
grouped_data
```

```python
plt.hist(df["rate"], bins = 5)
plt.title("Rating Distribution")
plt.show()
```

<b>Conclusion</b> : Majority Resturants received ratings from 3.5 to 4

```python
couple_data = df['approx_cost(for two people)']
sns.countplot(x = couple_data)
```

<b>Conclusion</b> : The majority couple prefer resturants with an approximate cost of 300 rupees

```python
plt.figure(figsize = (6, 6))
sns.boxplot(x = "online_order", y = "rate", data = df)
```

<b>Conclusion</b> : Offline orders received lower ratings in comparison to online orders, which obtained excellent ratings

```python
pivot_table = df.pivot_table(index = 'listed_in(type)', columns = 'online_order', aggfunc = "size", fill_value = 0)
sns.heatmap(pivot_table, annot = True, cmap = "YlGnBu", fmt = 'd')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed in (Type)")
plt.show()
```

<b>Conclusion</b> : Dining resturants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clients prefer to place orders in person at resturants, but prefer online ordering at cafes.
